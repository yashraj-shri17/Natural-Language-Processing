```python
import pandas as pd
import numpy as np
```

```python
messages = pd.read_csv(r"C:\Users\des\Desktop\ALL AI\Ai\spam.csv", encoding='latin1')
messages.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace = True,axis=1)
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^A-Za-z]',' ',messages['v2'][i]) ## to remove special characters
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word,pos='v') for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len([(i,j, k) for i, j, k in zip(map(len, corpus), corpus, messages['v2']) if i < 1]) ## these 8 are the messages that get lost in corpus creation because of specialized words its contiaining
## since some of the data points will be lost during training , we will have to make sure that Y variable at time of model training 
## should also have exactly same no of data points
```

```python
corpus
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess ## convert a document into list of lowercase tokens , ignoring tokens that are too short or too long
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
##Lets train our model from scratch
model = gensim.models.Word2Vec(words)
```

```python
def avg_word2vec(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if vectors:  # avoid empty list
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)  # or 300 if that's your vector size
```

```python
X = []
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
    
```

```python
## now the final indpendent feature
df = pd.DataFrame()
for i in range(0,len(X)):
    df = df.append(pd.DataFrame(X[i].reshape(1,-1)),ignore_index=True)
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier()
```

```python
classifier.fit(X_train,Y_train)
```

```python
Y_test_pred = classifier.predict(X_test)
Y_train_pred = classifier.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(Y_test,Y_test_pred),accuracy_score(Y_train,Y_train_pred)
```

```python
print(classification_report(Y_test,Y_test_pred))
```
