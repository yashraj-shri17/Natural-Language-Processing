```python
import pandas as pd
import numpy as np
```

```python
messages = pd.read_csv(r"C:\Users\des\Desktop\ALL AI\Ai\spam.csv", encoding='latin1')
```

```python
messages
```

```python
messages.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace = True,axis=1)
```

```python
messages.head()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^A-Za-z]',' ',messages['v2'][i]) ## to remove special characters
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word,pos='v') for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## create the Bag Of Words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000) ##taking 2000 words which are having the maximum frequency across the corpus
```

```python
##np.set_printoptions(threshold=np.inf) ## to print all the values of the array
X=cv.fit_transform(corpus).toarray()
```

```python
X ## this is with respect to , normal BOW , where we will getting count of the words
```

```python
##NOW FOR BINARY BOW
cv_binary = CountVectorizer(max_features = 200,binary=True) ## took less features just fir sake of example
```

```python
Y = cv_binary.fit_transform(corpus).toarray()
```

```python
Y
```

```python
## TO SEE THE TOP 200 WORDS
cv_binary.vocabulary_
```

### N-GRAMS

```python
## create the Bag Of Words with N-Grams
from sklearn.feature_extraction.text import CountVectorizer
cv_unigram = CountVectorizer(max_features=100,ngram_range=(1,1))##unigram - by default
cv_unigram.fit_transform(corpus).toarray()
cv_bigram = CountVectorizer(max_features=100,ngram_range=(2,2))##unigram to trigram
cv_bigram.fit_transform(corpus).toarray()
cv_trigram = CountVectorizer(max_features=100,ngram_range=(3,3))##bigram to trigram
cv_trigram.fit_transform(corpus).toarray()
```
